# # Computo valores faltantes

```python
data["home.dest"]=data["home.dest"].fillna("desconocido")
```

# # variables dummy

```python
dummy_sex = pd.get_dummies(data["sex"], prefix="sex")
```

```python
dummy_sex.head()
```

```python
column_name = data.columns.values.tolist()
```

```python
data = data.drop(["sex"], axis = 1)
```

```python
data = pd.concat([data, dummy_sex], axis=1)
```

# # Scatter plot

```python
figure, axs=plt.subplots(2,1,sharey=True, sharex=True)
data.plot(kind="scatter",x="age", y="survived", ax=axs[0])
data.plot(kind="scatter",x="age", y="pclass", ax=axs[1])
```

```python
data.plot(kind="scatter",x="age", y="sex_male")
```

```python
data.plot(kind="scatter",x="age", y="sex_female")
```

* La mayor parte de la población superior a 65 años es masculina y es de tercera clase, y no sobrevivió.
* La mayor parte de la población entorno a los 10 años es de tercera clase y no sobrevivió, y no hay variación de sexo.

# #Histograma de frecuencia

```python
plt.hist(data["age"], bins=[0,5,10,15,20,25,30,35,40,45,50,55,60,65,70,75,80])
plt.ylabel("frecuencia")
plt.xlabel("Edad")
plt.title("Histograma de edad")
```

```python
plt.hist(data["sex_female"], bins=[0,0.5,1])
plt.ylabel("frecuencia")
plt.xlabel("0 hombre,1 mujer")
plt.title("Mujer u hombre")
```

```python
plt.hist(data["survived"], bins=2)
plt.ylabel("frecuencia")
plt.xlabel("0 no, 1 sí")
plt.title("Supervivencia")
```

```python
plt.hist(data["pclass"], bins=3)
plt.ylabel("frecuencia")
plt.xlabel("clase; <2 primera, >2 y <3 segunda, 3 tercera")
plt.title("Clase")
```

* La cantidad de hombre duplica a la cantidad de mujeres.
* El menor rango de edad de personas es entre los 5 y 10 años, y entre una edad superior a los 65 años.
* Al ser menor el número de personas entre los 10 años y los superiores a los 65 años se entiende que no haya casi supervivientes entre esos rangos de edades.
* La cantidad de fallecidos es muy superior a la de supervivientes
* La mayor cantidad de pasajeros es de tercera clase.
